## class methods
The main use of class methods is defining methods that return an instance of the class, but aren't using the same code as __init__().

```python
class BetterDate:    
    # Constructor
    def __init__(self, year, month, day):
      # Recall that Python allows multiple variable assignments in one line
      self.year, self.month, self.day = year, month, day
    
    # Define a class method from_str
    @classmethod #always use this decorator before defining a class method
    def from_str(cls, datestr): #instead of self use cls
        # Split the string at "-" and convert each part to integer
        parts = datestr.split("-")
        year, month, day = int(parts[0]),int(parts[1]),int(parts[2]) #cast to int each index content of parts
        # Return the class instance
        return cls(year, month, day) #returning the class instance use cls before 
        
bd = BetterDate.from_str('2020-04-30')   
print(bd.year)
print(bd.month)
print(bd.day)
```

```python
# import datetime from datetime
from datetime import datetime

class BetterDate:
    def __init__(self, year, month, day):
      self.year, self.month, self.day = year, month, day
      
    @classmethod
    def from_str(cls, datestr):
        year, month, day = map(int, datestr.split("-"))
        return cls(year, month, day)
      
    # Define a class method from_datetime accepting a datetime object
    @classmethod
    def from_datetime(cls, datetime):
      return cls(datetime.year,datetime.month,datetime.day)

# You should be able to run the code below with no errors: 
today = datetime.today()     
bd = BetterDate.from_datetime(today)# Instance from a class and a class_method   
print(bd.year)
print(bd.month)
print(bd.day)
```

## String Representation

#### Instead of printing out the address location when we call print on an object, we use this two(__str() and __repr()) to print out a more informative output 

#### __str__() is used to print an informal output for the end user without the object type [1,2,3]
* only accepts self as the argument and should return a string

#### __repr__() is used to print out a formal output with the object type useful for the developers, outputs the reproducible information e.g Array([1,2,3])

#### Choose __repr__() since it acts as a fallback when __str__() is not defined 
* only accepts self and returns a string that can reproduce the object
* each data type should be represented as is e.g. a string should have quotes surrounding it
*  should return the expression that, when evaluated, will return the same object, ensuring the reproducibility of your code.

* refer to them by position index explicitly (like 'Insert {0} here'.format(num)) or

### Implementation of def \__str__()

```python
class Employee:
    def __init__(self, name, salary=30000):
        self.name, self.salary = name, salary
            
    # Add the __str__() method
    def __str__(self):
        emp = """
        Employee name: {name}
        Employee salary: {salary}
        """.format(name = self.name,\
                   salary = self.salary)
        return emp

emp1 = Employee("Amar Howard", 30000)
print(emp1)
emp2 = Employee("Carolyn Ramirez", 35000)
print(emp2)
```

### Implementation of def \__repr__()

```python
class Employee:
    def __init__(self, name, salary=30000):
        self.name, self.salary = name, salary
      

    def __str__(self):
        s = "Employee name: {name}\nEmployee salary: {salary}".format(name=self.name, salary=self.salary)      
        return s
      
    # Add the __repr__method  
    def __repr__(self):
            return 'Employee("{name}", {salary})'.format(name=self.name, salary = self.salary)

emp1 = Employee("Amar Howard", 30000)
print(repr(emp1))
emp2 = Employee("Carolyn Ramirez", 35000)
print(repr(emp2))
```

```python
# MODIFY the function to catch exceptions
def invert_at_index(x, ind):
    try:
        return 1/x[ind]
    except ZeroDivisionError:
        print("Cannot divide by zero!")
    except IndexError:
        print("Index out of range!")
        

 
a = [5,6,0,7]

# Works okay
print(invert_at_index(a, 1))

# Potential ZeroDivisionError
print(invert_at_index(a, 2))

# Potential IndexError
print(invert_at_index(a, 5))
```

### Implementing custom exceptions

```python
class SalaryError(ValueError): pass #Empty classes inherited from exception classes 
class BonusError(SalaryError): pass

class Employee:
    #class variables acting as global constants
  MIN_SALARY = 30000
  MAX_BONUS = 5000

  def __init__(self, name, salary = 30000):
    self.name = name    
    # dont assign it before the if 
    if salary < Employee.MIN_SALARY: #calling a class variable we use the class name
      raise SalaryError("Salary is too low!") # raise the exception  
    self.salary = salary
    
  # Rewrite using exceptions  
  def give_bonus(self, amount):
    if amount > Employee.MAX_BONUS:
       BonusError("The bonus amount is too high!")  
        
    elif self.salary + amount <  Employee.MIN_SALARY:
       SalaryError("The salary after bonus is too low!")
      
    else:  
      self.salary += amount
```

#### Notice that if you raise an exception inside an if statement, you don't need to add an else branch to run the rest of the code. Because raise terminates the function, the code after raise will only be executed if an exception did not occur.

#### It's better to include an except block for a child exception before the block for a parent exception, otherwise the child exceptions will be always be caught in the parent block, and the except block for the child will never be executed.

```python
emp = Employee("Katze Rik",\
                    50000)
try:
  emp.give_bonus(7000)
except BonusError: #child exception
  print("BonusError caught")
except SalaryError: #Parent exception
  print("SalaryError caught")
      
```

## Designing for inheritance and polymorphism

#### Polymorphism - using a unified interface to operate on objects of different classes
#### All that matters is the interface
#### Liskov Substitution Principle 
* Base class should be interchangeable with any of its subclasses without altering any properties of the program
* Syntatically - function signatures should be compatible both the arguments and the returned values
* Semantically - the state of the object and the program remains consistent 
        * subclass methods doesnt strengthen the inputs
        * subclass methods doesnt weaken the outputs
        * No additional exceptions
        
### No LSP no Inheritance


```python
class Parent:
    def talk(self):
        print("Parent talking!")     

class Child(Parent):
    def talk(self):
        print("Child talking!")          

class TalkativeChild(Parent):
    def talk(self):
        print("TalkativeChild talking!")
        Parent.talk(self)


p, c, tc = Parent(), Child(), TalkativeChild()

for obj in (p, c, tc):
    obj.talk()
```

## Managing data access: private attributes

#### In python all class data is public
#### Restricting Access
* You can use naming conventions
* use @property to customize access
* Overriding \__getattr__() and \__setattr__(), __repr__(),__str__()

#### Naming convention1: Internal Attributes(protected)
* start with an underscore
* obj.\_variablename, obj.\_method()
* The developer trust that the class users won't touch them
* Used for implementation details and helper functions

#### Naming convention2: Pseudo private attributes(private)
* starts with \__ and is the closest thing to private attr as used in other languages 
* Not inherited
* Name mangling - obj.\__attr_name is interpreted as obj.\_myclass__attr_name
* Used to prevent name clashes in child classes 
* Protects important attributes and methods that should not be overridden
* e.g a version attribute that stores the current version of the class and shouldn't be passed in the child class


### NB: Leading and trailing underscores are used for builtin python methods e.g \__init__()

```python
# Add class attributes for max number of days and months
class BetterDate:
    _MAX_DAYS = 30 #Internal Attributes
    _MAX_MONTHS = 12
    
    def __init__(self, year, month, day):
        self.year, self.month, self.day = year, month, day
        
    @classmethod
    def from_str(cls, datestr):
        year, month, day = map(int, datestr.split("-"))
        return cls(year, month, day)
    
    # Add _is_valid() checking day and month values
    #Internal method
    def _is_valid(self):
        #Make sure you call the class attributes with the class name
        if (self.day <= BetterDate._MAX_DAYS) and (self.month <= BetterDate._MAX_MONTHS):
            return True
        else:
            return False
    
bd1 = BetterDate(2020, 4, 30)
print(bd1._is_valid())

bd2 = BetterDate(2020, 6, 45)
print(bd2._is_valid())
```

## Properties - Allows to customie access

#### Customizing access can be like 
* Check the value for validity 
* Make attributes read only 

#### How to do this
* Use protected attribute with leading underscore to store data
* Use @property on a method whose name is the same as the name of the protected attribute then return the internal attribute
* Use @attr.setter on a method attr() that will be called on obj.attr = value
        * value to assign passed as an argument
* To create a read only property dont add @attr.setter
* To retrieve the property's value use @attr.getter
* To delete the property use @attr.deleter

#### Why are they useful
* Properties can be used to implement "read-only" attributes
* Properties control only one specific attribute that they're attached to. There are ways to prevent creating new attributes, but doing so would go against the "we're all adults here" principle.
* Properties can be accessed using the dot syntax just like regular attributes
* Properties allow for validation of values that are assigned to them

```python
class Customer:
    def __init__(self, name, new_bal):
        self.name = name
        if new_bal < 0:
           raise ValueError("Invalid balance!")
        self._balance = new_bal  

    # Add a decorated balance() method returning _balance        
    @property
    def balance(self):
        return self._balance

    # Add a setter balance() method
    @balance.setter
    def balance(self, new_bal):
        # Validate the parameter value
        if new_bal < 0:
           raise ValueError("Invalid balance!")
        self._balance = new_bal
        print("Setter method called")

# Create a Customer        
cust = Customer('Belinda Lutz', 2000)

# Assign 3000 to the balance property
cust.balance = 3000

# Print the balance property
print(cust.balance)
```
